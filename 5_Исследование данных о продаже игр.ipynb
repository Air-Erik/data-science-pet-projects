–ò–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –≤–∏–¥–Ω–æ, —á—Ç–æ –æ—Å–Ω–æ–≤–Ω–æ–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö —Å–æ–±—Ä–∞–Ω –∑–∞ –ø–µ—Ä–∏–æ–¥ –æ—Ç 2003 –¥–æ 2011. –¢–∞–∫ –∫–∞–∫ –Ω–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –±—É–¥—É—â–µ–µ —Ç–æ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 2003 –¥–æ 2016. –í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –º—ã –∑–∞—Ö–≤–∞—Ç–∏–º –æ—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤ –¥–∞–Ω–Ω—ã—Ö –∏ —É—á—Ç–µ–º –∞–∫—Ç—É–∞–ª—å–Ω—ã–µ —Ç—Ä–µ–Ω–¥—ã. –ò–Ω—Ç—Ä–µ–≤–∞–ª –∞–Ω–∞–ª–∏–∑–∞ –±—É–¥–µ—Ç —É—Ç–æ—á–Ω–µ–Ω –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º.

#### –ò–∑–º–µ–Ω–µ–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

–î–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ, –∞ —Ç–∞–∫–∂–µ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Å—Ä–æ–∫ –ø–æ—è–≤–ª–µ–Ω–∏—è –∏ –∏—Å—á–µ–∑–Ω–æ–≤–µ–Ω–∏—è –∏–≥—Ä–æ–≤—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º. –î–ª—è —ç—Ç–æ–≥–æ —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É. –í –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–æ–≤ –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –ø–ª–∞—Ç—Ñ–æ—Ä–º, –ø–æ–¥—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏. –î–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ä–æ–∫–∞ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –º–æ–∂–µ–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ –¥–∞—Ç—ã –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä –Ω–∞ —Ç–µ—Ö –∏–ª–∏ –∏–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö.

````python
platform_data = data.pivot_table(index='platform', 
                                 values=['sum_sales', 'year_of_release'], 
                                 aggfunc={'sum_sales': 'sum', 
                                          'year_of_release' : ['min', 'max']})
platform_data.columns = ['sum_sales', 'max_year', 'min_year']
platform_data
````

–ò—Å–∫–ª—é—á–∏–º –∏–∑ –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –ø—Ä–æ–¥–∞–∂–∞–º–∏ –º–µ–Ω–µ–µ 10 –º–ª–Ω –∫–æ–ø–∏–π. –¢–∞–∫ –∫–∞–∫ —ç—Ç–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –Ω–µ–ª—å–∑—è –Ω–∞–∑–≤–∞—Ç—å –º–∞—Å—Å–æ–≤—ã–º–∏ –∏ —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ —Ç–∞–∫–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –∫–æ—Ä–æ—Ç–∫–∏–π. –ò –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É –ø–æ —É–±—ã–≤–∞–Ω–∏—é

````python
platform_data = platform_data.query('sum_sales > 10')
platform_data = platform_data.sort_values(by='sum_sales', ascending=False)
````

–¢–æ–ø 12 –ø–ª–∞—Ç—Ñ–æ—Ä–º —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏

````python
index_pl = platform_data.index[:12]
index_pl = index_pl.to_list()
index_pl
````

##### –°—Ä–æ–∫ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

–î–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ä–æ–∫–∞ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤—ã—á—Ç–µ–º –∏–∑ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≥–æ–¥–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ. –ü–æ–ª—É—á–∏–º –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∂–∏–∑–Ω–∏ –∏–≥—Ä–æ–≤–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

````python
platform_data['live_years'] = (platform_data['max_year'] - platform_data['min_year'] + 1)
````
